```python
# ===== 1. Chọn tên cột brand =====
brand_col = "Please choose one of the business mobile applications Amazon Netflix Uber Lyft that you prefer to use."

# ===== 2. Chuẩn hoá: lower-case, xoá khoảng trắng =====
df[brand_col] = df[brand_col].astype(str).str.lower().str.strip()

# ===== 3. Sửa lỗi chính tả phổ biến =====
corrections = {
    "amzzon": "amazon",
    "amazone": "amazon",
    "amazn": "amazon",
    "amzon": "amazon",
    "uber ": "uber",
    " uber": "uber",
    "netflxi": "netflix",
    "netfllix": "netflix",
    "netlix": "netflix",
    "lyff": "lyft"
}
df[brand_col] = df[brand_col].replace(corrections)

# ===== 4. Hàm extract brand từ chuỗi dài =====
import re

def extract_brand(text):
    text = str(text).lower()

    # Nếu NA hoặc "nan" → Other
    if text in ["nan", "", "none"]:
        return "Other"

    # Bắt đúng brand hợp lệ
    if "amazon" in text:
        return "Amazon"
    if "netflix" in text:
        return "Netflix"
    if "uber" in text:
        return "Uber"
    if "lyft" in text:
        return "Lyft"

    # Nếu chuỗi chứa số dài hoặc không phải chữ
    if re.fullmatch(r"[0-9]+", text):
        return "Other"

    # Nếu chuỗi chứa list nhiều brand → lấy brand đầu tiên
    splits = re.split(r"[,\s/;]+", text)
    for s in splits:
        s = s.strip()
        if s in ["amazon", "netflix", "uber", "lyft"]:
            return s.capitalize()

    # Các app khác (flipkart, ebay, youtube…) → Other
    return "Other"

# ===== 5. Áp dụng ghi đè column gốc =====
df[brand_col] = df[brand_col].apply(extract_brand)

# ===== 6. Bảo đảm chỉ còn 5 nhóm =====
df[brand_col].value_counts()
```

```python
# --- BƯỚC 0: CÀI ĐẶT THƯ VIỆN CẦN THIẾT ---
!pip install factor_analyzer

import pandas as pd
import numpy as np
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
from factor_analyzer.factor_analyzer import calculate_kmo

# --- BƯỚC 1: HÀM TÍNH CRONBACH'S ALPHA CHI TIẾT ---
def cronbach_alpha_detailed(df, items):
    df_items = df[items]
    item_count = df_items.shape[1]
    correlation_matrix = df_items.corr()
    mean_correlation = correlation_matrix.values[np.triu_indices_from(correlation_matrix.values, k=1)].mean()
    alpha = (item_count * mean_correlation) / (1 + (item_count - 1) * mean_correlation)

    # Tính Item-Total Correlation cho từng biến
    item_total_corr = {}
    for col in items:
        # Tương quan giữa biến đó và tổng các biến còn lại
        other_items = [c for c in items if c != col]
        if len(other_items) > 0:
            total_others = df[other_items].sum(axis=1)
            corr = df[col].corr(total_others)
            item_total_corr[col] = corr
        else:
            item_total_corr[col] = 0.0

    return alpha, item_total_corr

# --- BƯỚC 2: KHAI BÁO CÁC NHÓM BIẾN (Mapping cũ của bạn) ---
# Lưu ý: Kiểm tra kỹ tên cột xem có khớp với df hiện tại không
scale_mapping = {
    'Mobility_Social': ['MLSS1', 'MLSS2'],
    'Mobility_Formal': ['MLSB1', 'MLSB2', 'MLSB3', 'MLSB4', 'MLSB5'],
    'Mobility_Personal': ['MLSP1', 'MLSP2', 'MLSP3'],
    'Experiential_Fit': ['Fpl', 'Fep', 'Fpe', 'Fv', 'InF', 'SeF', 'SyF', 'EmF'],
    'Identification': ['Iden', 'SeTrus'],
    'Citizenship_Behavior': ['Shr', 'Re', 'Ass', 'PrFeed', 'BaPi'],
    'CoCreation_Rewards': ['Meag', 'Coll', 'Con', 'Reg', 'Affe']
}

# --- BƯỚC 3: CHẠY CRONBACH'S ALPHA ---
print("\n" + "="*40)
print("PHẦN 1: KẾT QUẢ CRONBACH'S ALPHA")
print("="*40)

valid_items_for_efa = [] # List này sẽ gom các biến ĐẠT yêu cầu để chạy EFA sau đó

for name, items in scale_mapping.items():
    # Kiểm tra biến tồn tại
    cols = [c for c in items if c in df.columns]

    if len(cols) >= 2:
        alpha, item_corr = cronbach_alpha_detailed(df, cols)
        print(f"\n>>> Nhóm: {name}")
        print(f"    Cronbach's Alpha: {alpha:.4f}")

        # Check từng biến con
        print("    Item-Total Correlation:")
        good_items = []
        for item, corr in item_corr.items():
            status = "OK" if corr >= 0.3 else "LOẠI (< 0.3)"
            print(f"      - {item}: {corr:.4f} [{status}]")
            if corr >= 0.3:
                good_items.append(item)

        # Chỉ đưa các biến tốt vào list chạy EFA
        valid_items_for_efa.extend(good_items)
    else:
        print(f"\n>>> Nhóm {name}: Không đủ biến để chạy Alpha.")

# --- BƯỚC 4: CHẠY KMO & BARTLETT (EFA CHECK) ---
print("\n" + "="*40)
print("PHẦN 2: KMO & BARTLETT TEST (VALIDITY)")
print("="*40)

if len(valid_items_for_efa) > 0:
    df_efa = df[valid_items_for_efa].dropna()

    # 1. Bartlett
    chi_square_value, p_value = calculate_bartlett_sphericity(df_efa)
    print(f"Bartlett's Test P-value: {p_value:.4e}")
    if p_value < 0.05:
        print("=> ĐẠT (Các biến có tương quan với nhau)")
    else:
        print("=> KHÔNG ĐẠT (Sig > 0.05)")

    # 2. KMO
    kmo_all, kmo_model = calculate_kmo(df_efa)
    print(f"KMO Index: {kmo_model:.4f}")
    if kmo_model > 0.7:
        print("=> RẤT TỐT (> 0.7)")
    elif kmo_model > 0.5:
        print("=> CHẤP NHẬN ĐƯỢC (> 0.5)")
    else:
        print("=> KHÔNG ĐẠT (< 0.5)")
else:
    print("Không có biến nào đạt yêu cầu Reliability để chạy EFA.")
```

```python
# --- BƯỚC 1: CÀI ĐẶT THƯ VIỆN (Chỉ chạy 1 lần) ---
!pip install factor_analyzer

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
from factor_analyzer.factor_analyzer import calculate_kmo

# --- BƯỚC 2: CHỌN BIẾN ĐỂ CHẠY EFA ---
# Lưu ý: Mình đã loại 'SeTrus' (Alpha kém) và 'Iden' (biến đơn lẻ).
# Danh sách này dựa trên mapping bạn đã gửi

efa_cols = [
    # 1. Mobility - Social
    'MLSS1', 'MLSS2',

    # 2. Mobility - Formalistic
    'MLSB1', 'MLSB2', 'MLSB3', 'MLSB4', 'MLSB5',

    # 3. Mobility - Personalistic
    'MLSP1', 'MLSP2', 'MLSP3',

    # 4. Experiential Fit (Tất cả các biến Fit)
    'Fpl', 'Fep', 'Fpe', 'Fv', 'InF', 'SeF', 'SyF', 'EmF',

    # 5. Citizenship Behavior
    'Shr', 'Re', 'Ass', 'PrFeed', 'BaPi',

    # 6. Co-creation Rewards
    'Meag', 'Coll', 'Con', 'Reg', 'Affe'
]

# Tạo dataframe sạch, loại bỏ dòng trống
df_efa = df[efa_cols].dropna()

print(f"Số lượng quan sát đưa vào EFA: {len(df_efa)}")

# --- BƯỚC 3: KIỂM TRA ĐIỀU KIỆN (KMO & BARTLETT) ---
print("\n" + "="*40)
print("1. KIỂM ĐỊNH KMO VÀ BARTLETT")
print("="*40)

# Bartlett's Test
chi_square_value, p_value = calculate_bartlett_sphericity(df_efa)
print(f"Bartlett's Test Sig. (P-value): {p_value:.4e}")
if p_value < 0.05:
    print("=> ĐẠT: Các biến có mối tương quan với nhau.")
else:
    print("=> KHÔNG ĐẠT: Dữ liệu không phù hợp để phân tích nhân tố.")

# KMO Test
kmo_all, kmo_model = calculate_kmo(df_efa)
print(f"KMO Index: {kmo_model:.4f}")
if kmo_model > 0.7:
    print("=> RẤT TỐT (> 0.7): Dữ liệu cực kỳ phù hợp cho EFA.")
elif kmo_model > 0.5:
    print("=> CHẤP NHẬN ĐƯỢC (> 0.5).")
else:
    print("=> KHÔNG ĐẠT.")

# --- BƯỚC 4: QUYẾT ĐỊNH SỐ NHÂN TỐ (EIGENVALUES) ---
fa = FactorAnalyzer(rotation=None)
fa.fit(df_efa)
ev, v = fa.get_eigenvalues()

# Đếm số nhân tố có Eigenvalue > 1
n_factors = sum(ev > 1)
print(f"\n=> Số lượng nhân tố được trích xuất (Eigenvalue > 1): {n_factors}")

# Vẽ biểu đồ dốc sườn (Scree Plot)
plt.figure(figsize=(8, 4))
plt.plot(range(1, df_efa.shape[1]+1), ev, 'bo-')
plt.title('Scree Plot (Biểu đồ dốc sườn)')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.axhline(y=1, color='r', linestyle='--')
plt.grid()
plt.show()

# --- BƯỚC 5: CHẠY EFA CHÍNH THỨC (MA TRẬN XOAY) ---
# Sử dụng phương pháp trích Principal Component (method='principal')
# Phép xoay Varimax (rotation='varimax') để phân tách rõ các nhóm
fa_final = FactorAnalyzer(n_factors=n_factors, rotation='varimax', method='principal')
fa_final.fit(df_efa)

# Lấy bảng hệ số tải (Loadings)
loadings = pd.DataFrame(fa_final.loadings_, index=df_efa.columns)
loadings.columns = [f'Factor {i+1}' for i in range(n_factors)]

print("\n" + "="*40)
print("2. MA TRẬN XOAY (ROTATED COMPONENT MATRIX)")
print("(Đã ẩn các hệ số nhỏ < 0.4 để dễ nhìn cấu trúc)")
print("="*40)

# Hàm làm đẹp bảng: Ẩn số nhỏ < 0.4
def style_loading(val):
    return f"{val:.3f}" if abs(val) > 0.4 else "" # Để trống nếu < 0.4

# Hiển thị bảng
display(loadings.map(style_loading))

# --- BƯỚC 6: TỔNG PHƯƠNG SAI TRÍCH ---
variance = fa_final.get_factor_variance()
total_var = variance[2][-1] * 100 # Hàng cuối cùng (cumulative var), cột cuối cùng
print(f"\nTổng phương sai trích (Total Variance Explained): {total_var:.2f}%")
if total_var >= 50:
    print("=> ĐẠT: Các nhân tố giải thích được hơn 50% sự biến thiên của dữ liệu.")
else:
    print("=> THẤP: Mô hình chưa giải thích tốt dữ liệu.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats

# ==========================================
# BƯỚC 1: TẠO BIẾN TRUNG BÌNH (MEAN VARIABLES)
# ==========================================
# Dựa trên kết quả EFA "sạch" nhất mà chúng ta đã chốt:

# 1. MEAN MOBILITY: Gộp tất cả (Social + Formal + Personal) vì EFA gom chúng vào 1 Factor
# (Đã loại MLSB3 theo kết quả EFA trước đó)
mobility_cols = ['MLSS1', 'MLSS2', 'MLSB1', 'MLSB2', 'MLSB4', 'MLSB5', 'MLSP1', 'MLSP2', 'MLSP3']
df['Mean_Mobility'] = df[mobility_cols].mean(axis=1)

# 2. MEAN FIT: Gộp các biến Fit (Loại EmF vì cross-loading)
fit_cols = ['Fpl', 'Fep', 'Fpe', 'Fv', 'InF', 'SeF', 'SyF']
df['Mean_Fit'] = df[fit_cols].mean(axis=1)

# 3. MEAN REWARDS: Biến kết quả (Factor 3 EFA)
rewards_cols = ['Meag', 'Coll', 'Con', 'Reg', 'Affe']
df['Mean_Rewards'] = df[rewards_cols].mean(axis=1)

# 4. MEAN CITIZENSHIP: Giữ nguyên theo lý thuyết để làm biến phụ thuộc (DV) cuối cùng
citizenship_cols = ['Shr', 'Re', 'Ass', 'PrFeed', 'BaPi']
df['Mean_Citizenship'] = df[citizenship_cols].mean(axis=1)

# 5. MEAN IDENTIFICATION: Sử dụng biến đơn Iden (do SeTrus bị loại)
df['Mean_Identification'] = df['Iden']

print("✅ Đã tạo xong 5 biến đại diện để chạy mô hình!")
print(df[['Mean_Mobility', 'Mean_Fit', 'Mean_Rewards', 'Mean_Identification', 'Mean_Citizenship']].head())


# ==========================================
# BƯỚC 2: PHÂN TÍCH TƯƠNG QUAN (CORRELATION) - RQ2
# ==========================================
print("\n" + "="*50)
print("1. MA TRẬN TƯƠNG QUAN PEARSON (CORRELATION MATRIX)")
print("="*50)

# Chọn các biến chính
corr_vars = ['Mean_Mobility', 'Mean_Fit', 'Mean_Identification', 'Mean_Rewards', 'Mean_Citizenship']
corr_matrix = df[corr_vars].corr()

# Vẽ Heatmap cho đẹp (đưa vào bài báo cáo rất xịn)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".3f", linewidths=.5)
plt.title('Pearson Correlation Matrix')
plt.show()

# Hiển thị bảng số liệu
display(corr_matrix.round(3))


# ==========================================
# BƯỚC 3: HỒI QUY ĐA BIẾN (MULTIPLE REGRESSION) - RQ3, RQ4, RQ5
# ==========================================
print("\n" + "="*50)
print("2. KẾT QUẢ HỒI QUY (HYPOTHESIS TESTING)")
print("="*50)

# Hàm chạy hồi quy chuẩn APA
def run_regression_report(y_var, x_vars):
    # Chuẩn bị dữ liệu (bỏ NaN nếu có)
    data_reg = df[[y_var] + x_vars].dropna()
    Y = data_reg[y_var]
    X = data_reg[x_vars]
    X = sm.add_constant(X) # Thêm hằng số chặn (Intercept)

    model = sm.OLS(Y, X).fit()

    print(f"\n>>> MÔ HÌNH: {y_var} (Phụ thuộc) <-- {x_vars} (Độc lập)")
    print(f"   R-squared (Độ phù hợp mô hình): {model.rsquared:.3f} (Giải thích {model.rsquared*100:.1f}%)")
    print(f"   P-value (Độ tin cậy F-test): {model.f_pvalue:.4e}")

    # Tạo bảng kết quả đẹp
    summary_df = pd.DataFrame({
        'Hệ số Beta': model.params,
        'P-value': model.pvalues,
        'Kết luận': ['Chấp nhận (Có tác động)' if p < 0.05 else 'Bác bỏ (Không tác động)' for p in model.pvalues]
    })
    # Bỏ dòng const (hằng số) cho gọn khi xem
    display(summary_df.drop('const', errors='ignore').round(4))

# --- Chạy từng giả thuyết ---

# H1 (RQ3): Mobility tác động đến Fit?
run_regression_report('Mean_Fit', ['Mean_Mobility'])

# H2 & H3 (RQ4): Fit tác động đến Rewards và Identification?
run_regression_report('Mean_Rewards', ['Mean_Fit'])
run_regression_report('Mean_Identification', ['Mean_Fit'])

# H4 & H5 (RQ5): Rewards và Identification cùng tác động đến Citizenship?
run_regression_report('Mean_Citizenship', ['Mean_Rewards', 'Mean_Identification'])


# ==========================================
# BƯỚC 4: THỐNG KÊ SUY LUẬN SO SÁNH NHÓM (T-TEST & ANOVA)
# ==========================================
print("\n" + "="*50)
print("3. SO SÁNH NHÓM (T-TEST & ANOVA) - RQ6, RQ7")
print("="*50)

# RQ6: T-test - So sánh Fit giữa Nam và Nữ
# (Kiểm tra lại tên cột Gender trong df của bạn, ví dụ 'Gede')
if 'Gede' in df.columns:
    male = df[df['Gede'] == 'Male']['Mean_Fit']
    female = df[df['Gede'] == 'Female']['Mean_Fit']

    t_stat, p_val = stats.ttest_ind(male, female, equal_var=False) # Welch's t-test
    print(f"\n>>> Independent Sample T-Test (Gender vs Fit):")
    print(f"   Mean Male: {male.mean():.3f} | Mean Female: {female.mean():.3f}")
    print(f"   P-value: {p_val:.4f}")
    if p_val < 0.05:
        print("   => CÓ sự khác biệt có ý nghĩa thống kê.")
    else:
        print("   => KHÔNG có sự khác biệt (Nam/Nữ đánh giá Fit như nhau).")

# RQ7: ANOVA - So sánh Citizenship theo Học vấn (Edu)
if 'Edu' in df.columns:
    print(f"\n>>> One-way ANOVA (Education vs Citizenship):")
    # Lấy danh sách các nhóm học vấn
    groups = [df[df['Edu'] == g]['Mean_Citizenship'] for g in df['Edu'].unique() if pd.notna(g)]

    f_stat, p_val_anova = stats.f_oneway(*groups)
    print(f"   P-value: {p_val_anova:.4f}")
    if p_val_anova < 0.05:
        print("   => CÓ sự khác biệt giữa các nhóm học vấn.")
    else:
        print("   => KHÔNG có sự khác biệt.")
```
